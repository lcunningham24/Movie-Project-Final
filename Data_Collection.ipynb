# Webscraping IMDB
#### Top Box Office Movies from 2010-2019

```python
# list of urls to scrape 
num_list = [str(i) for i in np.arange(51,4952,50)]
def url():
    '''function that returns list of urls to webscrape'''
    url_list = ['https://www.imdb.com/search/title/?release_date=2010-01-01,2020-01-01&sort=boxoffice_gross_us,desc&view=advanced']
    for num in num_list: 
        url ='https://www.imdb.com/search/title/?release_date=2010-01-01,2020-01-01&sort=boxoffice_gross_us,desc&start='+num+'&ref_adv_nxt'
        url_list.append(url)
    return url_list 
url_list = url()
```

```python
# list of titles 
def movie_titles():
    title_list = []
    for url in url_list: 
        page = requests.get(url)
        soup = BeautifulSoup(page.content,'html.parser')
        page = soup.find('div',class_="lister-list")
        titles = [h3.find('a').get_text() for h3 in page.findAll('h3')]
        title_list.append(titles)
    return title_list
l = movie_titles()
```

```python
# turning list of lists into one list 
titles = []
for sublist in l:
    for item in sublist:
        titles.append(item)
titles
```

```python
# list of revenues 
revenue_list = []
for url in url_list: 
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    page = soup.find('div',class_="lister-list")
    for a in page.findAll('p',class_="sort-num_votes-visible"):
        x = a.findAll('span')[-1]
        revenue_list.append(x['data-value'])
```

```python
# adding a dollar sign to revenues
revenues = []
for revenue in revenue_list: 
    revenues.append('$'+revenue)
revenues
```

```python
# list of genres
genre_list = []
for url in url_list: 
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    page = soup.find('div',class_="lister-list")
    for genre in page.findAll('p',class_="text-muted"):
        genres = []
        genre = genre.find('span',class_="genre")
        genres.append(genre)
        for genre in genres: 
            if genre != None: 
                genre_list.append(genre.text.strip('\n').rstrip())
            else: 
                continue 
```

```python
# creating a df with title, revenue and genre 
df1 = pd.DataFrame([titles, revenues, genre_list]).transpose()
df1.columns = ['Movie Title','Gross Revenue (USD)','Genre']
df1.head()
```

```python
# filling in the 5 blank values at the end manually 
df1.at[4995,'Genre']='Comedy, Drama'
df1.at[4996,'Genre']='Comedy, Drama'
df1.at[4997,'Genre']='Drama'
df1.at[4998,'Genre']='Documentary'
df1.at[4999,'Genre']='Documentary'
```

# Webscraping Box Office Mojo

```python
# list of Box Office Mojo urls to webscrape 
def url():
    num_list = [str(i) for i in np.arange(2010,2021,1)]
    '''function that returns list of urls to webscrape'''
    url_list = []
    for num in num_list: 
        url = 'https://www.boxofficemojo.com/year/world/'+num
        url_list.append(url)
    return url_list 
url_list = url()
```

```python
# webscraping box office mojo for worldwide box office earnings to add to our dataframe
def title_boxoffice():
    y = {}
    for url in url_list:
        page = requests.get(url)
        soup = BeautifulSoup(page.content,'html.parser')
        page = soup.find('div',class_="a-section imdb-scroll-table-inner")
        for tr in page.findAll('tr'): 
            x = tr.find('a').get_text()
            if x == 'Rank':
                continue
            else:
                key = x
                val = (tr.find('td',class_='a-text-right mojo-field-type-money')).contents
                y[key]=val
    return y 
k = title_boxoffice() 

# turning dict into data_frame 
p = []
for ele in k.values():
    for i in ele: 
        p.append(i)
w = list(k.keys())

df_worldwide_boxoffice = pd.DataFrame([w, p]).transpose()
df_worldwide_boxoffice.columns = ['Movie Title','Worldwide Box Office (USD)']
df_worldwide_boxoffice

df_worldwide_boxoffice.to_csv('Worldwide_BoxOffice.csv')
```
